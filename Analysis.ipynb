### Analysis

This jupyter notebook will:
- examine clusters and analysis for the two unsupervised learning method


```python
# importing libraries and modules needed for clustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading model
model =  pd.read_pickle('model.pkl')
model
```

```python
# Interpreting K mean clusters
k_clusters = model.drop(['GMM CLUSTER'], axis=1)
k_clusters_pc_mean = k_clusters.groupby('KM CLUSTER').mean()
plt.figure(figsize=(15,15))
sns.heatmap(k_clusters_pc_mean, vmin=-1, vmax=1, center=0, cmap='vlag_r', square=True, annot=True)
```

```python
k_clusters[k_clusters['KM CLUSTER'] == 0].sort_values(by=['PC1', 'PC2', 'PC3', 'PC5', 'PC7'], ascending=False).head(20)
```

```python
k_clusters[k_clusters['KM CLUSTER'] == 1].sort_values(by=['PC5'], ascending=False).head(20)
```

```python
k_clusters[k_clusters['KM CLUSTER'] == 2].sort_values(by=['PC2', 'PC7', 'PC3', 'PC4', 'PC6'], ascending=False).head(20)
```

```python
# Interpreting GMM clusters
gmm_clusters = model.drop(['KM CLUSTER'], axis=1)
gmm_clusters_pc_mean = gmm_clusters.groupby('GMM CLUSTER').mean()
plt.figure(figsize=(15,15))
sns.heatmap(gmm_clusters_pc_mean, vmin=-1, vmax=1, center=0, cmap='vlag_r', square=True, annot=True)
```

```python
gmm_clusters[gmm_clusters['GMM CLUSTER'] == 0].sort_values(by=['PC3', 'PC5', 'PC6', 'PC1', 'PC7', 'PC4'], ascending=False).head(10)
```
